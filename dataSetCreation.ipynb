```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import csv
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
final_array = []
start = 2
end = 200

# For loop with defined start and end values
for i in range(start, end+1):
    url = 'https://www.primelocation.com/for-sale/property/west-yorkshire/leeds/?identifier=west-yorkshire%2Fleeds&page_size=50&q=Leeds%2C%20West%20Yorkshire&search_source=refine&radius=10&view_type=list&pn='+str(i)
    html = urlopen(url)
    soup = BeautifulSoup(html, 'lxml')
    type(soup)
    # Find all <li> elements with the desired class name
    li_elements = soup.find_all("li", class_="premium-listing")
    urlArray = []
    for li_element in li_elements:
        link_elements = li_element.find('a', class_='photo-hover')
        href = link_elements.get('href')
        urlArray.append(href)
        
    for value in urlArray:

        resultArray = []
        url = 'https://www.primelocation.com/'+value+'?search_identifier=8f6bd2e31818bc5482ae4bbe6fd8a14a'
        #url = "https://www.primelocation.com/for-sale/details/61660433/?search_identifier=02491891eb816c12c107d1e03a26f3dd"
        html = urlopen(url)
        soup = BeautifulSoup(html, 'lxml')
        type(soup)

        resultArray = [];

        segments = value.split("/")
        propertyId = segments[-2]
        resultArray.append(propertyId)

        houseTypeC = soup.find('h1', class_='listing-details-h1')
        houseTypeV = houseTypeC.text
        words = houseTypeV.split()
        houseTypeV = words[0]
        resultArray.append(houseTypeV)
        postcode = words[-1]
        resultArray.append(postcode)
        priceC = soup.find('span', class_='price')
        priceV = priceC.text.replace("\n", "")
        priceV = priceV.strip()
        resultArray.append(priceV)
        ntsArray = []
        elements = soup.find_all(class_="nts_value")
        num_elements = len(elements)
        for element in elements:
            ntsArray.append(element.text)

     
        tenure = ntsArray[0]    
        tenure = tenure.replace("\n", "")
        tenure = tenure.strip()

        
        if num_elements == 5:
            taxV = ntsArray[4]
        else:
            taxV = ntsArray[1]
        taxV = taxV.replace("\n", "")
        taxV = taxV.strip()

        resultArray.append(tenure)
        resultArray.append(taxV)

        bedElement = soup.find('span', class_='num-icon num-beds')
        if bedElement is not None:
            bed = bedElement['title']
            bedCount = bed.split()
            bedCount = bedCount[0]
        else:
            bedCount =0


        resultArray.append(bedCount)

        bathElement = soup.find('span', class_='num-icon num-baths')
        if bathElement is not None:
            bath = bathElement['title']
            bathCount = bath.split()
            bathCount = bathCount[0]
        else:
            bathCount =0
        resultArray.append(bathCount)

        receptionElement = soup.find('span', class_='num-icon num-reception')
        if receptionElement is not None:
            reception = receptionElement['title']
            receptionCount = reception.split()
            receptionCount = receptionCount[0]
        else:
            receptionCount =0
        resultArray.append(receptionCount)

        final_array.append(resultArray)
        #print(resultArray)

filename = 'DS_preminumRemaning.csv'

# Write data to CSV file
with open(filename, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(final_array)

print(f'CSV file "{filename}" created and downloaded.')
```
